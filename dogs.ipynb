```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("Dogs_life_expectancy.csv")
data.head()
```

```python
x = data["Adult weight males (kg)"]
y = data["Life expectancy (years)"]
```

```python
xc = sm.add_constant(x)
model = sm.OLS(y,xc)
result = model.fit()
ypred = result.predict(xc)
print(result.summary())
```

```python
plt.figure(dpi=1000)
plt.scatter(x,y,s=3)
plt.plot(x,ypred,c="purple")
plt.title("Dog breeds life expectancies as a function of their weight \n Y = 14 - 0.085×X    R² = 0.49    P-value < 0.000")
plt.xlabel("Average breed males weight (kg)")
plt.ylabel("Breed life expectancy (years)")
plt.savefig("Dogs_Life_expectancy")
plt.show()
```
